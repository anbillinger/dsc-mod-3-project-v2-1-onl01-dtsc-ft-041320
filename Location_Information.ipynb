```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns

data = pd.read_csv('data/Train_Values.csv')
target = pd.read_csv('data/Train_Labels.csv')
df = pd.concat([data,target['status_group']],axis=1)
```

```python
df = df.loc[df['longitude']!=0]
sns.scatterplot(df['longitude'],df['latitude'])
```

```python
func = df.loc[df['status_group']!='non functional']
non = df.loc[df['status_group']=='non functional']
```

```python
non_count = non['region'].value_counts()
all_count = df['region'].value_counts()
a_lst = []
for region in all_count.index:
    a_lst.append(non_count[region]/all_count[region])
temp_df = pd.DataFrame([all_count.index,a_lst]).transpose()
temp_df.columns = ['region','% non functional']
most_non_func = temp_df.sort_values('% non functional',ascending=False)[:10]
sns.barplot('region','% non functional',data=most_non_func)
plt.xticks(rotation=45)
plt.title('10 Regions with Highest Percent\nNon Functional Wells')
```
